```python
import numpy as np
import pandas as pd
# https://towardsdatascience.com/random-forest-in-python-24d0893d51c0
```

```python
df = pd.read_csv('clean_weather_data.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
df.head()
```

```python
df.loc[df['max_relative_humidity_v'] == 87.90]
```

```python
df.columns
```

```python
# Features and Target Data to Arrays
# the target variable is max_temperature_v
# we need to convert the pandas dataframe to numpy arrays because that's how the
# algorithm works

labels = np.array(df['max_temperature_v'])

# remove the labels from the features, axis = 1 refers to columns
df = df.drop('max_temperature_v', axis = 1)

# save feature names for later use
df_list = list(df.columns)

# convert to numpy array
features = np.array(df)
```

```python
# training and testing sets
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
# shape of data
# training features number of columns to match the testing feature number of columns 
# one to one matching

print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
train_features = np.nan_to_num(train_features)
train_labels = np.nan_to_num(train_labels)

rf.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data
test_features = np.nan_to_num(test_features)
predictions = rf.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Performance Metrics
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
