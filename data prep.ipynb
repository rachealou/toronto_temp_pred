```python
# split up date into year, month, and day columns
year = df['date'].apply(lambda x: x.split('-', 1)[0])
df['year'] = year
month = df['date'].apply(lambda x: x.split('-')[1])
df['month'] = month
day = df['date'].apply(lambda x: x.split('-')[2])
df['day'] = day
df.head()
```

```python
# filter for rows such that year = 2020 for training model
df = df.loc[df['year'] == '2020']
df.head()
```
